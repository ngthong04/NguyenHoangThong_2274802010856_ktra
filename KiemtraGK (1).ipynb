Bai 15

```python
# TÃ­nh chá» sá» BMI
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# PhÃ¢n loáº¡i BMI theo chuáº©n chÃ¢u Ã
def classify_bmi(bmi):
    if bmi < 18.5:
        return "Thiáº¿u cÃ¢n"
    elif 18.5 <= bmi <= 22.9:
        return "BÃ¬nh thÆ°á»ng"
    elif 23 <= bmi <= 24.9:
        return "Thá»«a cÃ¢n"
    else:
        return "BÃ©o phÃ¬"

# Ãp dá»¥ng phÃ¢n loáº¡i cho tá»«ng dÃ²ng
df['BMI_Category'] = df['cs_BMI'].apply(classify_bmi)

print(df)
```

Bai 16

```python
# TÃ­nh chá» sá» BMI
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# Chuáº©n hÃ³a cá»t <cs_BMI> vá» Äoáº¡n [0,1]
bmi_min = df['cs_BMI'].min()
bmi_max = df['cs_BMI'].max()
df['cs_BMI_scaled'] = (df['cs_BMI'] - bmi_min) / (bmi_max - bmi_min)

print(df)
```

Bai 17

```python
# Táº¡o cá»t vá»i tÃªn <GT>, Sau khi thá»±c hiá»n, xÃ³a cÃ¡c tá»« Mr, Mrs, Miss náº¿u trong cá»t <Name> cÃ²n cÃ¡c tá»« ÄÃ³.
df['GT'] = df['Name'].apply(lambda x: 'male' if 'Mr.' in x else 'female')
df['Name'] = df['Name'].str.replace('Mr. ', '').str.replace('Mrs. ', '').str.replace('Miss. ', '')
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dá»¯ liá»u ÄÃ£ cÃ³
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy",
             "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "GT": ["male", "female", "female", "female", "male", "male", "male",
           "female", "female", "female", "male", "female", "male", "male"],
    "BMI_Category": ["BÃ©o phÃ¬", "BÃ©o phÃ¬", "BÃ¬nh thÆ°á»ng", "Thá»«a cÃ¢n", "BÃ©o phÃ¬",
                     "BÃ©o phÃ¬", "BÃ©o phÃ¬", "BÃ©o phÃ¬", "BÃ¬nh thÆ°á»ng", "BÃ¬nh thÆ°á»ng",
                     "BÃ©o phÃ¬", "BÃ©o phÃ¬", "BÃ©o phÃ¬", "BÃ¬nh thÆ°á»ng"]
}

df = pd.DataFrame(data)

# Lá»c dá»¯ liá»u ngÆ°á»i ná»¯
female_data = df[df["GT"] == "female"]

# Äáº¿m sá» lÆ°á»£ng tá»«ng nhÃ³m trong <BMI_Category>
bmi_counts = female_data["BMI_Category"].value_counts()

# Váº½ biá»u Äá» cá»t
plt.figure(figsize=(6, 4))
plt.bar(bmi_counts.index, bmi_counts.values, color=['lightblue', 'pink', 'orange', 'red'])
plt.xlabel("PhÃ¢n Loáº¡i BMI")
plt.ylabel("Sá» lÆ°á»£ng")
plt.title("PhÃ¢n bá» BMI cá»§a ngÆ°á»i ná»¯")
plt.xticks(rotation=20)  # Xoay nhÃ£n trá»¥c x cho dá» nhÃ¬n
plt.show()
```
