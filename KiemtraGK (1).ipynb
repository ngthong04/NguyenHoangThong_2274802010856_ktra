Bai 15

```python
# Tính chỉ số BMI
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# Phân loại BMI theo chuẩn châu Á
def classify_bmi(bmi):
    if bmi < 18.5:
        return "Thiếu cân"
    elif 18.5 <= bmi <= 22.9:
        return "Bình thường"
    elif 23 <= bmi <= 24.9:
        return "Thừa cân"
    else:
        return "Béo phì"

# Áp dụng phân loại cho từng dòng
df['BMI_Category'] = df['cs_BMI'].apply(classify_bmi)

print(df)
```

Bai 16

```python
# Tính chỉ số BMI
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# Chuẩn hóa cột <cs_BMI> về đoạn [0,1]
bmi_min = df['cs_BMI'].min()
bmi_max = df['cs_BMI'].max()
df['cs_BMI_scaled'] = (df['cs_BMI'] - bmi_min) / (bmi_max - bmi_min)

print(df)
```

Bai 17

```python
# Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.
df['GT'] = df['Name'].apply(lambda x: 'male' if 'Mr.' in x else 'female')
df['Name'] = df['Name'].str.replace('Mr. ', '').str.replace('Mrs. ', '').str.replace('Miss. ', '')
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu đã có
data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy",
             "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Mr. William", "Miss. Hulda", "Mr. Lawrence", "Mr. Wilam"],
    "GT": ["male", "female", "female", "female", "male", "male", "male",
           "female", "female", "female", "male", "female", "male", "male"],
    "BMI_Category": ["Béo phì", "Béo phì", "Bình thường", "Thừa cân", "Béo phì",
                     "Béo phì", "Béo phì", "Béo phì", "Bình thường", "Bình thường",
                     "Béo phì", "Béo phì", "Béo phì", "Bình thường"]
}

df = pd.DataFrame(data)

# Lọc dữ liệu người nữ
female_data = df[df["GT"] == "female"]

# Đếm số lượng từng nhóm trong <BMI_Category>
bmi_counts = female_data["BMI_Category"].value_counts()

# Vẽ biểu đồ cột
plt.figure(figsize=(6, 4))
plt.bar(bmi_counts.index, bmi_counts.values, color=['lightblue', 'pink', 'orange', 'red'])
plt.xlabel("Phân Loại BMI")
plt.ylabel("Số lượng")
plt.title("Phân bố BMI của người nữ")
plt.xticks(rotation=20)  # Xoay nhãn trục x cho dễ nhìn
plt.show()
```
